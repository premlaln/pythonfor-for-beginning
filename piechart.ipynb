```python
#program for pie chart in details
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
number_of_votes=[2110,3505,1554,1287,127]
candidates_names=['Jaher Ali','Sofikul','Taher Ali','Mohammad wahid','Raziul']
explode=[0,0.1,0,0,0]
```

```python
plt.pie(number_of_votes,labels=candidates_names,explode=explode,shadow=False,startangle=0,autopct='%1.1f%%',wedgeprops={'edgecolor':'red'})
plt.title('Member of a union porisad selection')
plt.tight_layout()
plt.show()
```
